```python
# Review translations side by side
import json

for json_file in sorted(os.listdir(TRANSLATIONS_DIR)):
    if not json_file.endswith('.json'):
        continue
    path = os.path.join(TRANSLATIONS_DIR, json_file)
    with open(path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    print(f'\n=== {json_file} ({len(data)} segments) ===')
    for seg in data[:10]:
        print(f"  ZH: {seg['text_zh']}")
        print(f"  EN: {seg.get('text_en', seg.get('text_en_deepl', 'N/A'))}")
        print()
```
